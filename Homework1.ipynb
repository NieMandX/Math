```python
# define vector v1

v1 = [10, 30, 40]

# numpy version:
vector_length = np.sqrt(np.sum(np.array(v1) ** 2))

# python version:

def vec_length(vector):
    accum = 0
    for i in vector:
        accum += i**2
    return accum ** .5

print(vec_length(v1))
print(vector_length)
```

- ##### Задание 3.
> Напишите код на Python, реализующий построение графиков:
> - окружности,
> - эллипса,
> - гиперболы.




 - ##### **Окружность:**

```python
# ================================== Circle parameters =======================

# ------------------ Circle Radius -------------------------------------------
CircleR = 100
# ------------------ Circle center coordinates -------------------------------
BaseCircleX = -50
BaseCircleY = -50
# ------------------ Segments of a HalfCircle --------------------------------
HalfCircleSeg = 100

# ================================== X and Y Sequences equation ==============

# ------------------ X coordinates initial sequence --------------------------
CircleX = np.linspace(-CircleR + BaseCircleX, CircleR + BaseCircleX, HalfCircleSeg + 1)
# ------------------ Y coordinates sequence equation (Top part)---------------
CircleYTop = np.sqrt(CircleR ** 2 - (CircleX - BaseCircleX) ** 2) + BaseCircleY
# ------------------ Y coordinates sequence equation (Bottom part)------------
CircleYBottom = -np.sqrt(CircleR ** 2 - (CircleX - BaseCircleX) ** 2) + BaseCircleY

# ================================== Plot space initialization ===============
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plt.ylim(-CircleR * 1.2 + BaseCircleY, CircleR * 1.2 + BaseCircleY)
plt.xlim(-CircleR * 1.2 + BaseCircleX, CircleR * 1.2 + BaseCircleX)
plt.axhline(0, color = 'grey')
plt.axvline(0, color = 'grey')
plt.grid()
# ================================== Plotting the circle ===========

plt.plot(CircleX, CircleYTop, color = 'red')
plt.plot(CircleX, CircleYBottom, color = 'red');
```

 - ##### **Эллипс:**

```python
# ================================== Ellipse parameters =======================

# ------------------ Ellipse Radius -------------------------------------------
EllipseR1 = 50
EllipseR2 = 80

BigRadius = np.max([EllipseR1, EllipseR2])
SmallRadius = np.min([EllipseR1, EllipseR2])

# ------------------ Ellipse center coordinates -------------------------------
BaseEllipseX = 50
BaseEllipseY = -50
# ------------------ Segments of a HalfEllipse --------------------------------
HalfEllipseSeg = 300

# ================================== X and Y Sequences equation ==============

# ------------------ X coordinates initial sequence --------------------------
EllipseX = np.linspace(-BigRadius + BaseEllipseX, BigRadius + BaseEllipseX, HalfEllipseSeg + 1)
# ------------------ Y coordinates sequence equation (Top part)---------------
EllipseYTop = np.sqrt((1 - (((EllipseX - BaseEllipseX) ** 2) / BigRadius ** 2)) * SmallRadius ** 2)  + BaseEllipseY
# ------------------ Y coordinates sequence equation (Bottom part)------------
EllipseYBottom = -np.sqrt((1 - (((EllipseX - BaseEllipseX) ** 2) / BigRadius ** 2)) * SmallRadius ** 2)  + BaseEllipseY
# ================================== Plot space initialization ===============

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plt.ylim(-BigRadius * 1.2 + BaseEllipseY, BigRadius * 1.2 + BaseEllipseY)
plt.xlim(-BigRadius * 1.2 + BaseEllipseX, BigRadius * 1.2 + BaseEllipseX)
plt.axhline(0, color = 'grey')
plt.axvline(0, color = 'grey')
plt.grid()
# ================================== Plotting the Ellipse ===========

plt.plot(EllipseX, EllipseYTop, color = 'blue')
plt.plot(EllipseX, EllipseYBottom, color = 'blue');
```

```python
# ================================== Ellipse parameters =======================

# ------------------ Ellipse Radius -------------------------------------------
HyperR1 = 2
HyperR2 = 3

BigRadiusHyper = np.max([HyperR1, HyperR2])
SmallRadiusHyper = np.min([HyperR1, HyperR2])
dist = np.abs(HyperR1 - HyperR2)

# ------------------ Ellipse center coordinates -------------------------------
BaseHyperX = 0
BaseHyperY = 0
# ------------------ Segments of a HalfEllipse --------------------------------
HalfHyperSeg = 100

# ================================== X and Y Sequences equation ==============

# ------------------ X coordinates initial sequence --------------------------
HyperX = np.concatenate((np.linspace(SmallRadiusHyper - dist, SmallRadiusHyper, 200),
                         np.linspace(BigRadiusHyper, BigRadiusHyper + dist, 200)))
# ------------------ Y coordinates sequence equation (Top part)---------------
HyperYTop = np.sqrt((-1 + (((HyperX - BaseHyperX) ** 2) / HyperR1 ** 2)) * HyperR2 ** 2)  + BaseHyperY
# ------------------ Y coordinates sequence equation (Bottom part)------------
HyperYBottom = -np.sqrt((-1 + (((HyperX - BaseHyperX) ** 2) / HyperR1 ** 2)) * HyperR2 ** 2)  + BaseHyperY
# ================================== Plot space initialization ===============

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plt.ylim(-BigRadiusHyper * 1.2 + BaseHyperY, BigRadiusHyper * 1.2 + BaseHyperY)
plt.xlim(-BigRadiusHyper * 1.2 + BaseHyperX, BigRadiusHyper * 1.2 + BaseHyperX)
plt.axhline(0, color = 'grey')
plt.axvline(0, color = 'grey')
plt.grid()
# ================================== Plotting the Ellipse ===========

plt.plot(HyperX, HyperYTop, color = 'blue')
plt.plot(HyperX, HyperYBottom, color = 'blue');
```
